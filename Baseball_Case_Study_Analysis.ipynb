```python
#Spliting Data
x= df.drop('W', axis=1)
x
```

```python
y= df['W']
y
```

```python
# Standardization of Data

from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaled_x= scaler.fit_transform(x)
new_x= pd.DataFrame(data= scaled_x, columns= x.columns)
new_x
```

```python
x= new_x
x.shape
```

```python
y.shape
```

```python
#Machine Learning
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score, accuracy_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
#Finding Best Random State
maxACCU=0
maxRS=0
for i in range(1,500):
    x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=.3, random_state=i)
    lm= LinearRegression()
    lm.fit(x_train, y_train)
    pred_lm= lm.predict(x_test)
    acc= r2_score(y_test, pred_lm)
    if acc>maxACCU:
        maxACCU=acc
        maxRS= i
print(maxACCU)
print(maxRS)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=.3, random_state=114)
```

```python
lm= LinearRegression()
ls= Lasso()
rd= Ridge()
enr= ElasticNet()
ad= AdaBoostRegressor()
rf= RandomForestRegressor()

models= [lm, ls, rd, enr, ad, rf]

for m in models:
    print(m, '- R2_Score')
    m.fit(x_train, y_train)
    pred= m.predict(x_test)
    print(r2_score(y_test, pred))
    print()
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score

for m in models:
    scr_lm= cross_val_score(m, x,y, cv=5)
    print('CV score of ',m,  scr_lm.mean())
```

Observations: From above we can observe that for Lasso model there is less difference in r2_score and cross validation r2_score so we will select Lasso model for final predictions.

```python
#Hyperparameter Tuning for Best selected model LassoRegression

from sklearn.model_selection import GridSearchCV

parameters= {'alpha': [0.001,0.01,1],
    'selection': ['cyclic', 'random'],
    'normalize': [True, False]}

GSV= GridSearchCV(ls, parameters, cv=5)
GSV.fit(x_train, y_train)
GSV.best_params_
```

```python
GSV_predict= GSV.best_estimator_.predict(x_test)
print(r2_score(y_test, GSV_predict))
print('Mean absolute error:',mean_absolute_error(y_test, GSV_predict))
print('Mean squared error: ', mean_squared_error(y_test, GSV_predict))
print('Root mean squared error: ', np.sqrt(mean_squared_error(y_test, GSV_predict)))
```

```python
#Saving the Model
import joblib

joblib.dump(GSV.best_estimator_, 'base_ball_case_best_ls_model.obj')
```
